```python

# Örnek: df1 adında tüm kullanıcı event verisi var
df1["timestamp"] = pd.to_datetime(df1["timestamp"])
df1["date"] = df1["timestamp"].dt.date

#Her kullanıcı ve gün için sırayla yapılan eventleri listele
daily_patterns = (
    df1.sort_values("timestamp")
      .groupby(["user_id", "date"])["event_type"]
      .apply(list)
      .reset_index(name="daily_pattern")
)

user_daily_events_df = pd.DataFrame(daily_patterns)
user_daily_events_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Günlük ödeme sayısını hesapla
user_daily_events_df['daily_payment_count'] = user_daily_events_df['daily_pattern'].apply(lambda x: x.count('payment_success'))

# Yalnızca ödeme yapılan günleri filtrele
days_with_payment = user_daily_events_df[user_daily_events_df['daily_payment_count'] > 0]

# Kullanıcı bazlı toplam ödeme sayısı
total_payments = user_daily_events_df.groupby('user_id')['daily_payment_count'].sum().reset_index()

# Kullanıcı bazlı ödeme yapılan gün sayısı
payment_days = days_with_payment.groupby('user_id')['date'].nunique().reset_index()
payment_days = payment_days.rename(columns={'date': 'payment_days'})

# Birleştir
agg_df = total_payments.merge(payment_days, on='user_id', how='left')

# Eksik değer varsa sıfırla (her ihtimale karşı)
agg_df['payment_days'] = agg_df['payment_days'].fillna(0)

# Yeni oran: ödeme sayısı / ödeme yapılan gün sayısı
agg_df['payment_per_payment_day'] = agg_df['daily_payment_count'] / agg_df['payment_days']
agg_df['payment_per_payment_day'] = agg_df['payment_per_payment_day'].fillna(0)

# Normalize et
scaler = MinMaxScaler()
agg_df['normalized_payment_per_day'] = scaler.fit_transform(agg_df[['payment_per_payment_day']])

# Sonuç: kullanıcı, normalleştirilmiş oran ve ödeme yapılan gün sayısı
user_pay_day = agg_df[['user_id', 'normalized_payment_per_day', 'payment_days']]
user_pay_day = user_pay_day.rename(columns={'payment_days': 'active_days'})
user_pay_day = user_pay_day.sort_values(by='normalized_payment_per_day', ascending=False)

# Sonuçları göster
user_pay_day.head(20)
```

```python
import ast

# Sadece string olanları çevir
def parse_event_sequence(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except Exception as e:
            print(f"Hatalı satır: {x}\nHata: {e}")
            return []
    return x

df1["event_sequence"] = df1["event_sequence"].apply(parse_event_sequence)

# Prepare columns
first_event_type = []
last_event_before_first_payment = []
steps_before_first_payment = []
total_payment_success = []

# Process each row/user
for events in df1["event_sequence"]:
    # İlk olay
    first_event = events[0] if events else None

    # İlk ödeme öncesi olay ve kaç adım önce
    try:
        idx = events.index("payment_success")
        event_before = events[idx - 1] if idx > 0 else None
        steps = idx
    except ValueError:
        event_before = None
        steps = None

    # Toplam ödeme sayısı
    total_payments = events.count("payment_success")

    # Listeye ekle
    first_event_type.append(first_event)
    last_event_before_first_payment.append(event_before)
    steps_before_first_payment.append(steps)
    total_payment_success.append(total_payments)

# Yeni sütunlar olarak ata
df1["first_event_type"] = first_event_type
df1["last_event_before_first_payment"] = last_event_before_first_payment
df1["steps_before_first_payment"] = steps_before_first_payment
df1["total_payment_success"] = total_payment_success

df1.head()
```

```python
from collections import Counter
from nltk import ngrams

# "pre_payment_mod_repeat" sütununu eklemek için boş bir liste
pre_payment_mod_repeat = []

# Her kullanıcı için işlem yapıyoruz
for seq in df1["event_sequence"]:
    try:
        # İlk payment_success indeksini bul
        idx = seq.index("payment_success")

        # Sadece payment_success'ten önceki kısmı al
        pre_payment_seq = seq[:idx]

        # 2-gram üret (yeterli uzunluk varsa)
        if len(pre_payment_seq) >= 2:
            pre_payment_2grams = list(ngrams(pre_payment_seq, 2))

            # 2-gram'ların tekrar sayısını hesapla
            repeat_count = Counter(pre_payment_2grams)

            # En fazla tekrarlanan 2-gram sayısını al
            most_common_repeat_count = max(repeat_count.values()) if repeat_count else 0
        else:
            most_common_repeat_count = 0

    except ValueError:
        # payment_success yoksa atla
        most_common_repeat_count = 0

    # Sonuçları listeye ekle
    pre_payment_mod_repeat.append(most_common_repeat_count)

# Yeni sütunu DataFrame'e ekleyin
df1["pre_payment_mod_repeat"] = pre_payment_mod_repeat

# Sonuçları göster
df1.head()
```

```python
import pandas as pd

df1 = pd.read_csv("/home/eplinux/ikas_case_study/advanced_user_events_with_uuid.csv")

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Veri örneği: event_sequence'i liste formatından string'e çeviriyoruz
df = pd.read_csv("/home/eplinux/ikas_case_study/final_datasets/cluster_ready_data.csv")  # veya zaten bir DataFrame'in varsa bu satırı atla
df["event_sequence_str"] = df["event_sequence"].apply(lambda x: " ".join(eval(x)))

# TF-IDF vektörleştirme
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(df["event_sequence_str"])

# KMeans Clustering (örneğin 4 segment)
kmeans = KMeans(n_clusters=3, random_state=42)
df["behavior_cluster"] = kmeans.fit_predict(X_tfidf)

# (Opsiyonel) PCA ile 2D görselleştirme
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_tfidf.toarray())
plt.figure(figsize=(8, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=df["behavior_cluster"], cmap='viridis')
plt.title("Kullanıcı Davranış Kümeleme Görselleştirmesi")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()
```

```python
for i in range(4):  # 4 cluster olduğunu varsayalım
    print(f"\nCluster {i} için en yaygın olaylar:")
    cluster_users = df[df['behavior_cluster'] == i]
    all_events = " ".join(cluster_users['event_sequence_str'].tolist()).split()
    most_common = pd.Series(all_events).value_counts().head(5)
    print(most_common)
```

```python
# Cluster numaralarına karşılık gelen segment isimleri
cluster_labels = {
    0: "Growth Potential",
    1: "High Value Users",
    2: "Churn Risk"
}

# Yeni sütunu ekleme
df["behavior_segment"] = df["behavior_cluster"].map(cluster_labels)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from scipy.sparse import hstack

# --- Veri yükle ---
df1 = pd.read_csv("/home/eplinux/ikas_case_study/final_datasets/cluster_ready_data.csv")
df2 = pd.read_csv("/home/eplinux/ikas_case_study/advanced_user_profiles_with_uuid.csv")
pay_rf = pd.read_csv("/home/eplinux/ikas_case_study/new_feature_datas/PAY_rf_segmented_users.csv")
merged_df_ = pd.merge(df1, df2, on='user_id', how='outer')
merged_df_["segment_rf"] = pay_rf["segment"] + "_rf"

df = merged_df_.copy()

df["event_sequence_str"] = df["event_sequence"].apply(lambda x: " ".join(eval(x)))

# --- Özellikleri belirle ---
numeric_cols = ["total_payment_success", "purchase_intented_event_days_ratio", "steps_before_first_payment", "pre_payment_mod_repeat"]  # örnek sayısal sütunlar
categorical_cols = ["plan_type","country","device_type","industry", "segment_rf"]  # örnek kategorik sütunlar

# --- Sayısal + Kategorik Dönüştürme ---
preprocessor = ColumnTransformer([
    ("num", StandardScaler(), numeric_cols),
    ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols)
])

other_features = preprocessor.fit_transform(df)

# --- TF-IDF Vektörleştirme (event_sequence_str) ---
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df["event_sequence_str"])

# --- Hepsini Birleştir (sparse matrislerle çalış!) ---
X_all = hstack([tfidf_matrix, other_features])

# --- Kümeleme (örneğin KMeans) ---
kmeans = KMeans(n_clusters=3, random_state=42)
df["behavior_cluster"] = kmeans.fit_predict(X_all)

# --- Performans Ölçümü ---
score = silhouette_score(X_all, df["behavior_cluster"])
print("Silhouette Skoru (TF-IDF + Ek Özellikler):", round(score, 3))

# --- Görselleştirme (isteğe bağlı, PCA ile) ---
X_2d = PCA(n_components=2).fit_transform(X_all.toarray())
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=df["behavior_cluster"], cmap="viridis")
plt.title("Karma Özelliklerle Kümelenmiş Kullanıcılar")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()
```

```python
for i in range(4):  # 4 cluster olduğunu varsayalım
    print(f"\nCluster {i} için en yaygın olaylar:")
    cluster_users = df[df['behavior_cluster'] == i]
    all_events = " ".join(cluster_users['event_sequence_str'].tolist()).split()
    most_common = pd.Series(all_events).value_counts().head(5)
    print(most_common)
```
